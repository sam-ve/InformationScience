```python
import re
def onegrams(file):
    '''return the 100 most frequent words from a given file'''
    with open(file, 'r') as corpus:
        onegram_counts = {}                     #frequency dictionary
        for line in corpus.readlines():         #iterate over lines instead of complete file at once
            normalize = line.casefold()         #remove uppercase letters
            punct_re = re.compile(r'\W')        #split on non-word characters (remove punctuation and whitespace)
            words = punct_re.split(normalize)
            for word in words:
                if word in onegram_counts:      #compiling the frequency dictionary
                    onegram_counts[word] += 1
                else:
                    onegram_counts[word] = 1
        return sorted(onegram_counts.items(), key = lambda x: x[1], reverse=True)[1:101] #return 100 most frequent words (except for '')
```

```python
onegrams('corpus.txt')
```
